### 1. Generate 4 clusters

```python
#init test dataset
import numpy as np 

N = 80 
mean1 = [5, 15] 
mean2 = [5, 11] 
mean3 = [15,15] 
mean4 = [15,11] 
 
cov = [[3, 0], [0, 3]] 
 
np.random.seed(50) 
X = np.random.multivariate_normal( mean1, cov, int(N/4)) 
X = np.concatenate ((X, np.random.multivariate_normal ( mean2, cov, int(N/4)))) 
X = np.concatenate ((X, np.random.multivariate_normal ( mean3, cov, int(N/4)))) 
X = np.concatenate ((X, np.random.multivariate_normal ( mean4, cov, int(N/4))))
```

### 2. Implement K-means and Test 

```python
from numpy import * 
from matplotlib import pyplot as plt

def distance_euclidean(vector1, vector2):
    return sqrt(sum(power(vector1-vector2, 2)))  # return two vectors' distance

def rand_center(dataSet, k):
    n = np.shape(dataSet)[1]  # get the column number of dataset
    # init centroids，create matrix (k,n) fill with 0
    centroids = np.mat(zeros((k, n)))  # each centroid have n axis，total k centroids

    for j in range(n):
        # get min axis
        minJ = min(dataSet[:, j])
        # get range of axis
        rangeJ = float(max(dataSet[:, j]) - minJ)
        # from the min to max axis ,randomly generate a centroid
        centroids[:, j] = minJ + rangeJ * random.rand(k, 1)
    return centroids  


def k_means(dataSet, k, distMeas = distance_euclidean, creatCent = rand_center):
    m = np.shape(dataSet)[0] # get the row number of dataset
    # generate cluster assignment matrix，first column store the belonging centroid，
    # second column is the distance of point to corresponding centroid
    clusterAssment = np.mat(zeros((m, 2)))
    centroids = creatCent(dataSet, k) 
    
    # determining whether clustering has converged
    clusterChanged = True
    while clusterChanged:
        clusterChanged = False
        for j in range(50):# set max iteration as 50
            for i in range(m):  # let point to its nearest centroid
                minDist = inf 
                minIndex = -1 #init
                for j in range(k):
                    # calculate the distance of each point to the new corresponding centroid
                    distJI = distMeas(centroids[j,:],dataSet[i,:])
                    if distJI < minDist:
                        # if point i to centroid j is closer, assignment i to j
                        minDist = distJI
                        minIndex = j          
                if clusterAssment[i,0] != minIndex:
                    clusterChanged = True
                # store the cluster assinment result
                clusterAssment[i,:] = minIndex,minDist**2
        for cent in range(k):  # recalculate the centroid
            ptsInClust = dataSet[nonzero(clusterAssment[:, 0].A == cent)[0]]
            centroids[cent, :] = mean(ptsInClust, axis=0)
            
    #calculate SSE
    sse = 0 #init SSE
    for i in range(m):
        sse = sse + clusterAssment[i,1] #sum clusterAssment stored minDist**2 get SSE
        
    return centroids, clusterAssment, sse
    
def plt_kmeans():
    datMat = X
    myCentroids, clusterAssing, sse = k_means(datMat, 4)
    #plot dataset
    plt.scatter(array(datMat)[:, 0], array(datMat)[:, 1], c=array(clusterAssing)[:, 0].T)  
    #plot centroid
    plt.scatter(myCentroids[:, 0].tolist(), myCentroids[:, 1].tolist(), c="r") 
    plt.show()
    print("SSE:", sse)
```

### 3. Implement K-means++ and Test 

```python
from numpy import *
from matplotlib import pyplot as plt

def distance_euclidean(vector1, vector2):
    return sqrt(sum(power(vector1-vector2, 2)))  # return two vectors' distance

def nearest(point, centroids):
    min_dist = inf
    m = np.shape(centroids)[0]  # the number of centroids
    for i in range(m):
        # each distance of every point to every centroids
        d = distance_euclidean(point, centroids[i, ])
        # select min_dist
        if min_dist > d:
            min_dist = d
    return min_dist

def get_center(dataSet, k):
    m, n = np.shape(dataSet)
    centroids = np.zeros((k , n))
    index = np.random.randint(0, m)
    centroids[0,] = dataSet[index, ]
    # init a distance range()
    d = [0.0 for _ in range(m)]
    for i in range(1, k):
        sum_all = 0
        for j in range(m):
            # find the nearset centroid to each point
            d[j] = nearest(dataSet[j, ], centroids[0:i, ])
            # sum all the min_distance
            sum_all += d[j]
        sum_all *= random.rand()
        # get the farest point as centroid
        for j, di in enumerate(d):
            sum_all=sum_all - di
            if sum_all > 0:
                continue
            centroids[i,] = dataSet[j, ]
            break
    return centroids

def k_means_plus(dataSet,k,distMeas = distance_euclidean,creatCent = get_center):

    m = np.shape(dataSet)[0] # get the row number of dataset
    # generate cluster assignment matrix，first column store the belonging centroid，
    # second column is the distance of point to corresponding centroid
    clusterAssment = np.mat(zeros((m, 2)))
    centroids = creatCent(dataSet, k) 
    
    # determining whether clustering has converged
    clusterChanged = True
    while clusterChanged:
        clusterChanged = False
        for j in range(50):# set max iteration as 50
            for i in range(m):  # let point to its nearest centroid
                minDist = inf 
                minIndex = -1 #init
                for j in range(k):
                    # calculate the distance of each point to the new corresponding centroid
                    distJI = distMeas(centroids[j,:],dataSet[i,:])
                    if distJI < minDist:
                        # if point i to centroid j is closer, assignment i to j
                        minDist = distJI
                        minIndex = j          
                if clusterAssment[i,0] != minIndex:
                    clusterChanged = True
                # store the cluster assinment result
                clusterAssment[i,:] = minIndex,minDist**2
        for cent in range(k):  # recalculate the centroid
            ptsInClust = dataSet[nonzero(clusterAssment[:, 0].A == cent)[0]]
            centroids[cent, :] = mean(ptsInClust, axis=0)
            
    #calculate SSE
    sse = 0 #init SSE
    for i in range(m):
        sse = sse + clusterAssment[i,1] #sum clusterAssment stored minDist**2 get SSE
        
    return centroids, clusterAssment, sse

def plt_kmeans_plus():
    datMat = X
    myCentroids, clusterAssing, sse = k_means(datMat, 4)
    #plot dataset
    plt.scatter(array(datMat)[:, 0], array(datMat)[:, 1], c=array(clusterAssing)[:, 0].T)  
    #plot centroid
    plt.scatter(myCentroids[:, 0].tolist(), myCentroids[:, 1].tolist(), c="r") 
    plt.show()
    print("SSE:", sse)

```

I test both K-means and K-means++ five times.

Qualitatively, Both of the two method can get good and bad clustering result.And the good result is almost the same. The clustering result of K-means always change, and the result of K-means++ is more stable. The clusters looks have a more similar size in generated picture which SSE is better.

Quantitatively, In total 10 times tests, the best SSE of the clustering result is 343.33, and the worst SSE is 561.52. Besides，I got 1 good result in 5 tests in K-means while K-means++ got 4 times good performance in 5 tests, so K-means is more stable to get good clustering result.

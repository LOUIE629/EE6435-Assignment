```python
import numpy as np 

#Euclidean Distence
def Euclidean(A, B):
    euclidean_temp = np.sqrt(sum(np.power((A - B), 2))) #求Euclidean的公式
    euclidean = round(euclidean_temp, 2) #保留两位小数
    return euclidean

#Cosine Similarity
def Cos(A, B):
    A_norm = np.linalg.norm(A) 
    B_norm = np.linalg.norm(B) #求向量的范数，get ||A||，||B|| 
    cos = np.dot(A, B)/(A_norm * B_norm) #求Cos的公式：<A,B>/||A||||B||
    return cos

#Correlation
def Corr(A, B):
    cov_ab = np.cov(A, B)[0][1]
    A_std = np.sqrt(sum(np.power(A - np.mean(A),2))/(np.size(A)-1))
    B_std = np.sqrt(sum(np.power(B - np.mean(B),2))/(np.size(B)-1))
    if A_std * B_std == 0:
        return "0/0(undefined)" #分母为0时无意义，返回undefined
    else:
        corr_temp = cov_ab/(A_std * B_std) #求Correlation的公式
        corr = round(corr_temp, 2) #保留两位小数
        return corr

#Jaccard
def Jaccard(n, A, B):
    A = np.array(A)
    B = np.array(B)
    n = int(n)
    if (A != 0 | 1).all() or (B != 0 | 1).all():
        return "undefined" #当输入数组不为0，1数组时返回undefined
    f11 = 0
    f00 = 0 
    for i in range(0, n):  #遍历两个数组，得到f11和f00的值
        if A[i] == 1 and B[i] == 1:
            f11 = f11 + 1
        elif A[i] == 0 and B[i] == 0:
            f00 = f00 + 1 
    jaccard = f11 / (n - f00) #求Jaccard的公式:J = f11 / （f01 + f10 + f11）
    return jaccard
                    
def main():
    n = input()
    X_input = input()
    Y_input = input() #按要求输入数据
    X_list = list(map(int,X_input.split()))
    Y_list = list(map(int,Y_input.split())) #将输入的字符串转为int型数组
    X = np.array(X_list)
    Y = np.array(Y_list) #使用np.array，来使用相关numpy函数
    return print("output:""\n""cos(x,y) = ",Cos(X,Y),"\n""corr(x,y) = ",Corr(X,Y),"\n""Euclidean(x,y) = ",Euclidean(X,Y),"\n""Jaccard(x,y) = ",Jaccard(n, X, Y))
    #if __name__ == '__main__':
    #    main()
```
